## 환경설정

```python
from bs4 import BeautifulSoup
import requests

def get_html(url):
    response = requests.get(url)
    html = BeautifulSoup(response.text, 'html.parser')
    return html

#print(get_html('https://news.kbs.co.kr/news/pc/main/main.html'))
```

## KBS News

```python
url = 'https://news.kbs.co.kr/news/pc/main/main.html'
html = get_html(url)

news_titles = html.select('div.head-line p.title')
len(news_titles)

titles = list() 

for news_title in news_titles:
    text = news_title.text.strip()
    titles.append(text)

titles
```

- list comprehension 사용


```python
titles = [ x.text.strip() for x in news_titles ]
titles
```

## Quotes 수집

```python
html = get_html('https://quotes.toscrape.com/')

quote_elements = html.select('.quote')
len(quote_elements)

quotes = list() # quotes = []

for quote_element in quote_elements:

    # 1. 데이터 추출하기
    quote_text = quote_element.select_one('.text').text
    author = quote_element.select_one('.author').text

    # 2. 테그 추출하기
    tag_elements = quote_element.select('.tag')
    tags = [ x.text for x in tag_elements ]
    
    '''
    tags = list()
    for tag_element in tag_elements:
        tag = tag_element.text
        tags.append(tag)
    '''
    
    # 3. 명언과 저자를 dict 로 만들기
    quote = dict() # quote = {}
    quote['text'] = quote_text
    quote['author'] = author
    quote['tags'] = tags

    # 4. 리스트에 그 결과를 넣기
    quotes.append(quote)

print(len(quotes))
quotes
```
