## 실습 15

- driver launch

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# find_elements 사용을 위해 필요
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

def find(css_selector):
    return driver.find_element(By.CSS_SELECTOR, css_selector)

def finds(css_selector):
    return driver.find_elements(By.CSS_SELECTOR, css_selector)

def element_find(element, css_selector):
    return element.find_element(By.CSS_SELECTOR, css_selector)

def element_finds(element, css_selector):
    return element.find_elements(By.CSS_SELECTOR, css_selector)

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.implicitly_wait(10)  
```

- 날짜별로 승률 테이블 가져오기

```python
import time 
# 팀 순위 페이지 접속
driver.get('https://www.koreabaseball.com/Record/TeamRank/TeamRankDaily.aspx')


# 날짜별 승률을 저장하기 위한 리스트
win_rates = list()

while(True):

    # 1. 2025년도 데이터만 추출하기 위헤 날짜 확인
    
    # 1-1. 날짜 추출 
    date = find('span#cphContents_cphContents_cphContents_lblSearchDateTitle').text
    
    # 1-2. 년도 확인하여 2025년이 아니면 종료
    year = int(date.split('.')[0])  # 날짜에서 연도만 추출
    if year != 2025:
        break

    print(f'데이터 수집 - {date}')
    
    
    # 2. 페이지에서 팀명과 승률을 수집하여 win_rates에 저장
    records = finds('#cphContents_cphContents_cphContents_udpRecord > table.tData > tbody > tr')
    win_rate = dict()
    for record in records:
        fields = element_finds(record, 'td')

        #팀명 추출, 2번째 열
        name = fields[1].text
        #승률, 7번째 열
        rate = float(fields[6].text)

        # 수집한 데이터 저장
        #print(name, rate)
        win_rate[name] = rate

    # 날짜 오름차순으로 저장하기 위해 앞에다 삽입
    win_rates.insert(0, win_rate)
    
    # 3. 이전 날짜로 이동하기
    
    # 3-1. 이전 페이지 버튼 찾기
    prev = find('input#cphContents_cphContents_cphContents_btnPreDate')
    prev.click()

    # 3-2. click 후 1초긴 대기, 대기 시간이 없으면, 페이지가 바뀌기 전에 데이터를 추출하게됨
    time.sleep(1)
    
print(win_rates)
```

- 승률 데이터로 부터 ndarray 생성

```python

teams = None
result = None

for win_rate in win_rates:
    
    # win_rate dict를 팀명으로 정렬
    win_rate = sorted(win_rate.items())

    # 정렬된 값에서 승률값으로 ndarray 생성
    teams = [ key for key, value in win_rate ] 
    win_rate = [ value for key, value in win_rate ]
    win_rate = np.array(win_rate)

    # 생성된 array를 열기준으로 합치기
    if result is None:
        result = win_rate
    else:
        result = np.column_stack((result, win_rate))

print(teams)
print(result)
```

- 파일로 저장하기

```python
with open('kbo_win_rate.csv', 'w') as f:
    for i, win_rate in enumerate(result):
        team = teams[i]
        win_rate = ','.join(win_rate.astype(str))
        line = team + ',' + win_rate + '\n'
        f.write(line)
        
```

```python
import requests
from bs4 import BeautifulSoup
import numpy as np

# 0. 시작 url
url = "https://www.koreabaseball.com/Record/TeamRank/TeamRankDaily.aspx"

response = requests.get(url)
html = BeautifulSoup(response.text, 'html.parser')

table = html.select_one('#cphContents_cphContents_cphContents_udpRecord > table.tData')
records = html.select('#cphContents_cphContents_cphContents_udpRecord > table.tData > tbody > tr')
rows = list()
for record in records:

    # record에서 필드값을 갖는 요소들을 찾기
    fields = record.select('td')
    
    row = list()
    # 각 요소에서 text 추출
    for field in fields:
        row.append(field.text.strip())

    rows.append(row)

rows[0]

arr = np.array(rows)
print(arr)
```
