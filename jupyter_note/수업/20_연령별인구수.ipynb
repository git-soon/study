# 행정 기관 열에서 마지막 단어만 추출

```python
# 1. 행정 기관에 스페이스가 있어 제거
temp = pop_df['행정기관'].str.strip()
temp1 = pop_df['행정기관'].str
temp2 = temp1.strip()
print(temp2)
#print(temp)

# 2. 스페이스로 스플릿
temp = temp.str.split(' ')
#print(temp)

# 3. 마지막 단어 선택
temp = temp.str[-1]
#print(temp)
```

# search 열에 행정구역 마지막 단어 삽입

```python
pop_df['search'] = temp
pop_df.head()
```

# search 열을 조건으로 필터링
