```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
tips.head()
#size 는 인원수
```

```python
sns.scatterplot(
    x= 'total_bill',
    y='tip',
    alpha=0.8,
    data=tips
)
```

```python
sns.scatterplot(
    x= 'total_bill',
    y='tip',
    style='sex',
    hue='smoker',
    alpha=0.8,
    data=tips
)
# 스타일과  hue 는 각 1개씩만 지정할수 있다
```

```python
#실습 1 
#total_bill 과 tip의 비율 (몇프로를 팁으로 내는지)의 관계를 시각화 하고, 남여를 다른 색으로 표시하시요

tips['tip_rate'] = (tips['tip']/tips['total_bill'])


sns.scatterplot(
    y= 'tip_rate',
    x='total_bill',
    hue='sex',
    alpha=0.8,
    data=tips
)
```

```python
# stripplot : 단순 데이터 분포를 빠르게 보고 싶을 때
#  예시; 요일별 tip 분포 확인

sns.stripplot(
    x='day',
    y='tip',
    jitter=False,
    data=tips
)
```

```python
#실습 2 
 # 점심/저녁 별 tip의 분포를 시각화 하시요.
 # 흡연자/비흡연자를 색깔로 구분

sns.stripplot(
    x='time',
    y='tip',
    hue = 'sex',
    jitter = True,
    dodge=True,
    data=tips
)
```

```python
#catplot

sns.catplot(
    x='day',
    y='tip',
    hue='sex',
    kind='box',
    data=tips
)
```
