## 실습 2 3개 파일 주식 가격 데이터 그래프 그리기

```python
#내가 처음한것, 데이터에 개행문자(\n) 있어서 그래프 이상하게 나옴
#개행문자는 매우 작은 숫자로 표기된다.

with open ("./데이터/APPL.txt") as A:
    contents_A = A.readlines()

with open ("./데이터/MSFT.txt") as M:
    contents_M = M.readlines()

with open ("./데이터/XOM.txt") as X:
    contents_X = X.readlines()

A_prices = [float (i) for i in contents_A]
M_prices = [float (i) for i in contents_M]
X_prices = [float (i) for i in contents_X]

plt.plot(contents_A, label='APPL')
plt.plot(contents_M, label='MSFT')
plt.plot(contents_X, label='XOM')
plt.legend()

#집와서 다시보니 개행문자가 나타난게 price를 안하고 contents를 해서 그런것이다.
#생각해보면 for 에서 개행문자가 다 날라갈텐데 ㅋㅋㅋ;;;
```

```python
# print(contents_A)
# print(A_prices) #개행 문자 없어진건 strip이 아니라 그냥 for 때문임
# 그러면 그냥 contents.str.replace('\n','')도 되나?
contents.str.replace('\n','') # 안된다. 오브젝트가 아니라 str도 안되고 for문으로 \n을 다 없애는거하는거 빼곤 그냥 위처럼 하는게 낫다
```

```python
# 수정 : 값들 보니 다 개행문자 포함되어 있어서, 개행문자를 float 하기 전에 지워주고 들어가야한다.

with open ("./데이터/APPL.txt") as A:
    contents_A = A.readlines()

with open ("./데이터/MSFT.txt") as M:
    contents_M = M.readlines()

with open ("./데이터/XOM.txt") as X:
    contents_X = X.readlines()

A_prices = [float (i.strip()) for i in contents_A] # i 가 contendts 를 순회하며 값을 가져오는것 이니 i에 strip()을 해서 개행문자를 없애면 돼야한다
M_prices = [float (i.strip()) for i in contents_M]
X_prices = [float (i.strip()) for i in contents_X]

plt.plot(A_prices, label='APPL')
plt.plot(M_prices, label='MSFT')
plt.plot(X_prices, label='XOM')
# plt.legend(loc = 'lower center',title='3 graphs') 
plt.title("3 graph")
#legend( loc : 태그위치, title : 태그 타이틀)
#.title("그래프 이름") 이거랑 legned(title) 헷갈리지 말자
#여러 데이터 하나의 그래프에 그리기 -> 한 셀에서 plot()을 여러 번 호출 ->plot()마다 label 값을 지정하고, legend를 표시라고 하는데 legend없어도 여러개 그려지긴함
```

### 실습 3 두 리스트를 점선과 마크를 사용하여 그려라

```python
google = [40, 35, 10, 15, 50, 60, 70, 50, 30]
microsoft = [10, 20, 40, 60, 58, 55, 75, 80, 100]

plt.title("Two graph with dots")
plt.plot(google, label='google',color = 'red', linestyle = '--', marker="o", markerfacecolor = 'red')   #라벨 왜 출력 안됨? = ㄹ및에 legend 없었을때는 라벨 출력 안됐음
plt.plot(microsoft,label = 'microsoft', color = 'blue', linestyle = '-', marker="s", markerfacecolor = 'blue') # 라인 스타일과 도트 는 다양하다
plt.xlabel("time(week)", fontsize=15)
plt.ylabel("price(USD)", fontsize=15)
plt.legend(title='3 graphs')
```

```python
with open ("./데이터/APPL.txt") as A:
    contents_A = A.readlines()

with open ("./데이터/MSFT.txt") as M:
    contents_M = M.readlines()

with open ("./데이터/XOM.txt") as X:
    contents_X = X.readlines()

A_prices = [float (i.strip()) for i in contents_A] 
M_prices = [float (i.strip()) for i in contents_M]
X_prices = [float (i.strip()) for i in contents_X]

plt.subplot(2,2,1)
plt.plot(A_prices, label='APPL', color = "red")
plt.title('APPL')

plt.subplot(2,2,2)         
plt.plot(M_prices, label='MSFT', color = "blue")
plt.title('MSFT')

plt.subplot(2,2,3)
plt.plot(X_prices, label='XOM', color = "green")
plt.title('XOM')

plt.xlabel("time(week)")
plt.ylabel("price(USD)")
plt.tight_layout() # 간격 조정

plt.suptitle("3 graphs") # 전체 그래프의 제목
```

### 실습5  주사위를 20번 던져서 나온 값의 빈도수를 히스토그램으로 그려라

```python
import random as rd

dice = []
for i in range(20):
    dice.append(rd.randint(1,6))

plt.hist(dice, bins = 6, histtype = 'bar')
```
